**Visualization**

In this section, we create and analyze visualizations to better understand the patterns and trends in the data.

**Distribution of Flight Delays (DepDelay and ArrDelay)**

Purpose: The histograms visualize the distribution of flight delays, focusing on both departure delays (DepDelay) and arrival delays (ArrDelay). By filtering out negative delay values, the plots show the frequency and severity of delays in the flight data.

Axes:

- X-axis: Represents the delay time in minutes, starting from 0 minutes to higher delay values.
- Y-axis: Represents the frequency, indicating how many flights experienced delays for each delay range.

Key Features:

- Departure Delay Histogram: Shows the distribution of delays for flights at departure.
- Arrival Delay Histogram: Shows the distribution of delays for flights at arrival.

We found that there were negative arrival and departure delay records, so we excluded those from our analysis of delay distribution.

```python
# Check for negative values in DepDelay and ArrDelay
negative_dep_delays = df[df['DepDelay'] < 0].shape[0]
negative_arr_delays = df[df['ArrDelay'] < 0].shape[0]

print(f"Number of negative departure delays: {negative_dep_delays}")
print(f"Number of negative arrival delays: {negative_arr_delays}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Filter out negative delays for both DepDelay and ArrDelay
df_filtered = df[(df['DepDelay'] >= 0) & (df['ArrDelay'] >= 0)]

# Plot distribution of departure delays
plt.figure(figsize=(12, 6))
sns.histplot(df_filtered['DepDelay'].dropna(), kde=True, color='blue', bins=50)
plt.title('Distribution of Departure Delays')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Frequency')
plt.xlim(left=0)  # Set x-axis to start at zero
plt.ylim(bottom=0)  # Set y-axis to start at zero
plt.show()

# Plot distribution of arrival delays
plt.figure(figsize=(12, 6))
sns.histplot(df_filtered['ArrDelay'].dropna(), kde=True, color='red', bins=50)
plt.title('Distribution of Arrival Delays')
plt.xlabel('Arrival Delay (minutes)')
plt.ylabel('Frequency')
plt.xlim(left=0)  # Set x-axis to start at zero
plt.ylim(bottom=0)  # Set y-axis to start at zero
plt.show()
```

**Scatter Plot Showing Carrier Delay vs. Departure Delay**

Purpose: The scatter plot visualizes the relationship between the CarrierDelay (delays caused by the airline carrier) and the DepDelay (overall departure delay). This allows us to observe how delays caused by the carrier influence or correlate with overall departure delays.

Axis:

- X-axis: Represents the CarrierDelay in minutes, which shows the delay caused by the airline carrier.

- Y-axis: Represents the DepDelay in minutes, showing the overall departure delay of the flight.

```python
# Scatter plot of Carrier Delay vs. Departure Delay
plt.figure(figsize=(12, 6))
sns.scatterplot(x='CarrierDelay', y='DepDelay', data=df, alpha=0.6)
plt.title('Carrier Delay vs. Departure Delay')
plt.xlabel('Carrier Delay (minutes)')
plt.ylabel('Departure Delay (minutes)')
plt.show()
```

**Boxplot Showing Airline vs. Departure Delay**

Purpose: The boxplot compares the departure delays across different airlines (Reporting_Airline). It shows the distribution of departure delays for each airline.

Axis:

- X-axis: Represents the different Reporting_Airline codes (airlines).
- Y-axis: Represents the DepDelay in minutes, showing the departure delay times for each airline.

```python
# Boxplot of different airlines and their average departure delays
plt.figure(figsize=(12, 6))
sns.boxplot(x='Reporting_Airline', y='DepDelay', data=df)
plt.title('Airline vs. Departure Delay')
plt.xlabel('Airline')
plt.ylabel('Departure Delay (minutes)')
plt.xticks(rotation=90)
plt.show()
```

**Line Plot Showing Weather Delay vs. Departure Delay**

Purpose: This line plot visualizes the relationship between weather delays (WeatherDelay) and departure delays (DepDelay). The plot shows how changes in weather delays might impact departure delays, with each data point representing a flight's weather delay and corresponding departure delay.

Axis:

- X-axis: Represents the WeatherDelay in minutes, showing the amount of delay caused by weather conditions.
- Y-axis: Represents the DepDelay in minutes, showing the corresponding departure delay for each flight.

```python
# Line plot for weather delay vs. departure delay (grouped by time)
plt.figure(figsize=(12, 6))
sns.lineplot(x='WeatherDelay', y='DepDelay', data=df, marker='o', color='green')
plt.title('Weather Delay vs. Departure Delay')
plt.xlabel('Weather Delay (minutes)')
plt.ylabel('Departure Delay (minutes)')
plt.show()
```

**Line Plot Showing Average Monthly Departure Delay** 

Purpose: The plot shows the average departure delay for each month, helping to identify any trends or patterns in delays throughout the year.

Axis:

- X-axis: Month
- Y-axis: Average Departure Delay (minutes)

```python
# Line plot for monthly delay trends (avg DepDelay)
monthly_avg_delay = df.groupby('Month')['DepDelay'].mean()
plt.figure(figsize=(12, 6))
monthly_avg_delay.plot(kind='line', marker='o', color='purple')
plt.title('Average Departure Delay by Month')
plt.xlabel('Month')
plt.ylabel('Average Departure Delay (minutes)')
plt.xticks(range(1, 13))
plt.show()
```

**Boxplot Showing Departure Delay by Day of the Week**

Purpose: The boxplot displays the distribution of departure delays for each day of the week, highlighting variations and potential outliers in delays for different days.

Axis:

- X-axis: Day of the Week
- Y-axis: Departure Delay (minutes)

```python
# Boxplot for flight delays by day of the week
plt.figure(figsize=(12, 6))
sns.boxplot(x='DayOfWeek', y='DepDelay', data=df)
plt.title('Departure Delay by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Departure Delay (minutes)')
plt.show()
```

**Line Plot Showing Average Flight Delay by Year**

Purpose: This line plot shows the trend of average departure delays over multiple years, allowing for an understanding of whether delays have increased or decreased over time.

Axis:

- X-axis: Year
- Y-axis: Average Departure Delay (minutes)

```python
# Group data by Year and calculate the average Departure Delay per year
df['Year'] = pd.to_datetime(df['FlightDate']).dt.year  # Ensure the year column is in datetime format
avg_delay_per_year = df.groupby('Year')['DepDelay'].mean()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(avg_delay_per_year.index, avg_delay_per_year.values, marker='o', color='b', linestyle='-', linewidth=2)
plt.title('Average Flight Delay by Year')
plt.xlabel('Year')
plt.ylabel('Average Departure Delay (minutes)')
plt.grid(True)
plt.show()
```

**Scatter Plots Showing Flight Distance vs Departure Delay & Flight Distance vs Arrival Delay**

Purpose: These scatter plots examine how flight distance correlates with both departure and arrival delays, helping identify if longer flights tend to have more delays.

Axis:

- X-axis: Flight Distance (miles)
- Y-axis: Departure Delay (minutes) for the first plot
- Y-axis: Arrival Delay (minutes) for the second plot

```python
# Scatter plot for Distance vs Departure Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Distance', y='DepDelay', color='blue', alpha=0.6)
plt.title('Flight Distance vs Departure Delay')
plt.xlabel('Flight Distance (miles)')
plt.ylabel('Departure Delay (minutes)')
plt.grid(True)
plt.show()

# Scatter plot for Distance vs Arrival Delay
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Distance', y='ArrDelay', color='red', alpha=0.6)
plt.title('Flight Distance vs Arrival Delay')
plt.xlabel('Flight Distance (miles)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True)
plt.show()
```

**Line Plots Showing Departure Delay vs. Time of Day & Arrival Delay vs. Time of Day**

Purpose: These line plots show the average departure and arrival delays based on the hour of the day, providing insights into peak times when delays are more likely.

Axis:

- X-axis: Hour of the Day (ranging from 0 to 23)
- Y-axis: Average Departure Delay (minutes) for the first plot
- Y-axis: Average Arrival Delay (minutes) for the second plot

```python
# Convert DepTime and ArrTime to hours for better visualization
df['DepHour'] = df['DepTime'] // 100  # Extract hour from DepTime
df['ArrHour'] = df['ArrTime'] // 100  # Extract hour from ArrTime

# Average Departure Delay by Hour
avg_dep_delay_by_hour = df.groupby('DepHour')['DepDelay'].mean()

# Average Arrival Delay by Hour
avg_arr_delay_by_hour = df.groupby('ArrHour')['ArrDelay'].mean()

# Plotting Departure Delay vs. Time of Day (DepTime)
plt.figure(figsize=(10, 6))
plt.plot(avg_dep_delay_by_hour.index, avg_dep_delay_by_hour.values, marker='o', color='blue', label='Departure Delay')
plt.title('Departure Delay vs. Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Departure Delay (minutes)')
plt.grid(True)
plt.legend()
plt.show()

# Plotting Arrival Delay vs. Time of Day (ArrTime)
plt.figure(figsize=(10, 6))
plt.plot(avg_arr_delay_by_hour.index, avg_arr_delay_by_hour.values, marker='o', color='red', label='Arrival Delay')
plt.title('Arrival Delay vs. Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Arrival Delay (minutes)')
plt.grid(True)
plt.legend()
plt.show()
```
